```python
import os
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
NARX_PATH = os.path.join(os.getcwd(), 'NARX')
NL_PATH = os.path.join(os.getcwd(), 'Non-Linear')
RNN_path = os.path.join(os.getcwd(), 'RNN')
```

```python
NARX_train = pd.read_csv(os.path.join(NARX_PATH, 'train_NARX_5.csv'), index_col=0)
NARX_val = pd.read_csv(os.path.join(NARX_PATH, 'val_NARX_5.csv'), index_col=0)
```

```python
NL_train = pd.read_csv(os.path.join(NL_PATH, 'train_NonLinear_5.csv'), index_col=0)
NL_val = pd.read_csv(os.path.join(NL_PATH, 'val_NonLinear_5.csv'), index_col=0)
```

```python
RNN_res = pd.read_csv(os.path.join(RNN_path, 'RNN.csv'), index_col=0, header=None)
RNN_res.rename(columns={1: 'train', 2: 'val'}, inplace=True)
```

```python
#plt.plot(NARX_train.groupby("epoch").mean(), label='NARX_train')
plt.plot(NARX_val['loss'], label='NARX_val')
#plt.plot(NL_train.groupby("epoch").mean(), label='NL_train')
plt.plot(NL_val['loss'], label='NL_val')
#plt.plot(RNN_res['train'], label='RNN_train')
plt.plot(RNN_res['val'], label='RNN_val')

plt.xlabel("epoch")
plt.ylabel("loss")

plt.yscale('log')
plt.legend()
plt.show()
```
